Step 2: drop the row we do not need for this project.

```python
df = df.drop('UGDS', axis = 1)
df
```

Step 3: remove all rows with NaN and "PrivacySuppressed"

```python
df = df[df.SAT_AVG_ALL.notnull()]
df = df[df.UGDS_RICH.notnull()]
df = df[df.MD_EARN_WNE_P10.notnull()]
df
```

```python
df = df[~df.MD_EARN_WNE_P10.str.contains('PrivacySuppressed')]
```

```python
df.shape
```

Step 4: rename rows for easy access.

```python
school = df['INSTNM'].values
SAT = df['SAT_AVG_ALL'].values
family = df['UGDS_RICH'].values
income = df['MD_EARN_WNE_P10'].values
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    y = family,
    x = income,
    text = school,
    mode = 'markers',
    line = dict(
        color = 'purple',
        width = 4
    )
)

)
fig.update_layout(title='income as function of family income',
                   xaxis_title='Family Income',
                   yaxis_title='Student Income'
                 )

fig.show()
```

Step 5: 2D Plot. More information about 2D plotting can be found here: https://plotly.com/python/line-charts/

```python
import plotly.graph_objs as go
import chart_studio.plotly as py
```

Step 6: 3D plot. More information can be found here: https://plotly.com/python/3d-scatter-plots/ 

```python

data = [go.Scatter3d(
        x = SAT,
        y = family,
        z = income,
        mode='markers',
        marker = dict(
            size = 4,
            color = family,
            colorscale = 'Viridis',
            opacity = 0.8
        )
       

)]
layout = go.Layout(
         margin= dict(
         l = 5, 
         r = 5,
         t = 5,
         b = 0
         
         )

)
fig2 = go.Figure( data = data, layout = layout)
fig2.show()
```
